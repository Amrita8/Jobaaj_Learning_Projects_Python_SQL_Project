ADVANCED
PROBLEMS:

1. Calculate the moving average of order values for each customer over their order history.

```python
# 1 Moving avg order value
if merged is not None:
  val=merged.groupby('order_id')['revenue'].sum().reset_index()
  val=val.merge(orders[['order_id','customer_id','order_purchase_timestamp']],on='order_id')
  val=val.sort_values(['customer_id','order_purchase_timestamp'])
  print(val.groupby(['customer_id','order_id','order_purchase_timestamp'])['revenue'].rolling(3,min_periods=1).mean())
```

2. Calculate the cumulative sales per month for each year.

```python
# 2 Cumulative monthly sales
if merged is not None:
  monthly_revenue = merged.groupby(merged['order_purchase_timestamp'].dt.to_period('M'))['revenue'].sum().reset_index()
  monthly_revenue['order_year'] = monthly_revenue['order_purchase_timestamp'].dt.year
  monthly_revenue['order_month'] = monthly_revenue['order_purchase_timestamp'].dt.month
  monthly_revenue['Cumulative revenue'] = monthly_revenue.groupby('order_year')['revenue'].cumsum()
  monthly_revenue = monthly_revenue[['order_year', 'order_month', 'revenue', 'Cumulative revenue']]
  monthly_revenue.columns = ['order_year', 'order_month', 'Monthly revenue', 'Cumulative revenue']
  print(monthly_revenue)
```

```python
# 12 Cumulative monthly sales
if merged is not None:
    m=merged.groupby(merged['order_purchase_timestamp'].dt.to_period('M'))['revenue'].sum()
    # Create a MultiIndex from the year and month of the PeriodIndex
    m.index = pd.MultiIndex.from_arrays([m.index.year, m.index.month])
    cum = m.groupby(level=0).cumsum().unstack(level=0, fill_value=0)
    cum.plot(figsize=(5,5), title="Cumulative Monthly Sales per Year")
    plt.show()
```

3. Calculate the year-over-year growth rate of total sales.

```python
# 3 YoY growth
if merged is not None:
  y = merged.groupby(merged['order_purchase_timestamp'].dt.year)['revenue'].sum().reset_index()
  y['YoY growth rate'] = y['revenue'].pct_change()*100
  y.columns = ['order_year', 'total_sales', 'YoY growth rate']
  print(y)
```

```python
if merged is not None:
    y=merged.groupby(merged['order_purchase_timestamp'].dt.year)['revenue'].sum()
    growth=y.pct_change()*100
    growth.plot(kind='bar', figsize=(5,3), title="YoY Growth %")
    plt.show()
```

4. Calculate the retention rate of customers, defined as the percentage of customers who make another purchase within 6 months of their first purchase.

```python
# 4 Retention rate
if orders is not None:
  f=orders.groupby('customer_id')['order_purchase_timestamp'].min()
  sub=orders.merge(f,on='customer_id')
  sub['days']=(sub['order_purchase_timestamp_x']-sub['order_purchase_timestamp_y']).dt.days
  retained=sub[(sub['days']>0)&(sub['days']<=180)]['customer_id'].nunique()
  print(retained/len(f)*100)
```

5. Identify the top 3 customers who spent the most money in each year.

```python
# 15 Top 3 customers per year
if merged is not None:
  t=merged.groupby([merged['order_purchase_timestamp'].dt.year,'customer_id'])['revenue'].sum().reset_index()
  t=t.rename(columns={'order_purchase_timestamp':'order_year','revenue':'total_spent'})
  t['spending_rank']=t.groupby('order_year')['total_spent'].rank(method='dense',ascending=False)
  top3=t[t['spending_rank']<=3].sort_values(['order_year','spending_rank'])
  print(top3)
```

```python
if merged is not None:
    t = merged.groupby([merged['order_purchase_timestamp'].dt.year, 'customer_id'])['revenue'].sum().reset_index()
    t = t.rename(columns={'order_purchase_timestamp':'order_year','revenue':'total_spent'})
    t['spending_rank'] = t.groupby('order_year')['total_spent'].rank(method='dense', ascending=False)
    top3 = t[t['spending_rank'] <= 3]

    for year, grp in top3.groupby('order_year'):
        plt.figure(figsize=(6,5))   # 👈 set figure size here
        plt.bar(grp['customer_id'], grp['total_spent'], color="skyblue")
        plt.title(f"Top 3 Customers in {year}")
        plt.xlabel("Customer ID")
        plt.ylabel("Total Spent")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()
```
